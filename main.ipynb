Data Preprocessing

```python
df_ml = df.copy()
```

```python
df_ml.drop("customerID", axis=1, inplace=True)
```

```python
df_ml["TotalCharges"] = pd.to_numeric(
    df_ml["TotalCharges"],
    errors="coerce"
)

df_ml["TotalCharges"] = df_ml["TotalCharges"].fillna(
    df_ml["TotalCharges"].median()
)
```

```python
print("Missing Values:")
print(df_ml.isnull().sum())
```

```python
df_ml["TotalCharges"].fillna(df_ml["TotalCharges"].median(), inplace=True)
```

```python
df_ml["Churn"] = df_ml["Churn"].map({"No": 0, "Yes": 1})
```

```python
df_ml = pd.get_dummies(df_ml, drop_first=True)
```

```python
X = df_ml.drop("Churn", axis=1)
y = df_ml["Churn"]
```

```python
print("\nFeature Matrix Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
df_ml["TotalCharges"] = pd.to_numeric(df_ml["TotalCharges"], errors="coerce")

df_ml["TotalCharges"] = df_ml["TotalCharges"].fillna(df_ml["TotalCharges"].median())
```

```python
print(df_ml.isnull().sum())
```

```python
display(df_ml.head())
```

Train-Test Split & Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42,stratify=y)
```

```python
print("Training Features Shape :", X_train.shape)
print("Testing Features Shape  :", X_test.shape)
print("Training Target Shape   :", y_train.shape)
print("Testing Target Shape    :", y_test.shape)
```

```python
logistic_model = LogisticRegression(max_iter=1000)

logistic_model.fit(X_train,y_train)

logistic_predictions = logistic_model.predict(X_test)
```

```python
print("Missing values in X_train:")
print(X_train.isnull().sum()[X_train.isnull().sum() > 0])

print("\nTotal missing values:")
print(X_train.isnull().sum().sum())
```

```python
logistic_accuracy = accuracy_score(y_test,logistic_predictions)

logistic_precision = precision_score(y_test,logistic_predictions)

logistic_recall = recall_score(y_test,logistic_predictions)

logistic_f1 = f1_score(y_test,logistic_predictions)
```

```python
print("Accuracy:", logistic_accuracy)
print("Precision:", logistic_precision)
print("Recall:", logistic_recall)
print("F1 Score:", logistic_f1)
```

```python
print("\nClassification Report:")
print(classification_report(y_test,logistic_predictions))
```

Decision Tree Model

```python
tree_model = DecisionTreeClassifier(random_state=42)
```

```python
tree_model.fit(X_train,y_train)
```

```python
tree_predictions = tree_model.predict(X_test)
```

```python
tree_accuracy = accuracy_score(y_test,tree_predictions)
```

```python
tree_precision = precision_score(y_test,tree_predictions)
```

```python
tree_recall = recall_score(y_test,tree_predictions)
```

```python
tree_f1 = f1_score(y_test,tree_predictions)
```

```python
print("Accuracy:", tree_accuracy)
print("Precision:", tree_precision)
print("Recall:", tree_recall)
print("F1 Score:", tree_f1)
```

```python
print("\nClassification Report:")
print(
    classification_report(
        y_test,
        tree_predictions
    )
)
```

```python
comparison = pd.DataFrame(
    {"Model": ["Logistic Regression","Decision Tree"],
        "Accuracy": [logistic_accuracy,tree_accuracy],
        "Precision": [logistic_precision,tree_precision],
        "Recall": [logistic_recall,tree_recall],
        "F1 Score": [logistic_f1,tree_f1]
    }
)
```

```python
display(comparison)
```

```python
best_model = comparison.loc[comparison["F1 Score"].idxmax()]

print("\nBest Performing Model:")
print(best_model)
```
